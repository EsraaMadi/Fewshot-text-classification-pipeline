```python
import pickle
import yaml
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
#general set up 
config_path = "../config.yml"
data_name = 'Restaurant_Reviews'
```

```python
# read score file
ymlfile = open(config_path)
cfg = yaml.load(ymlfile, Loader=yaml.FullLoader)
score_path = cfg[data_name]['supervised-model-score-path']
file = open(score_path ,'rb')
scores = pickle.load(file)
scores['score'] = scores['score']*100
scores
```

```python
# print text examples
file = open('../models/Restaurant_Reviews/Supervised_model/test.pkl' ,'rb')
test_df = pickle.load(file)
test_df[test_df['pred_labels_id']!= test_df['actual_label_id']]
```

```python
test_df
```

```python
# prepare data for visulization 
id_columns = ['id', 'train_set', 'test_set', 'creation_date']
scores_long = pd.melt(scores,
                      id_vars = id_columns,
                      value_vars=[i for i in scores.columns if i not in id_columns],
                      value_name='score', ignore_index=False)
scores_long#.head(2)
```

```python
scores
```

```python
fig = px.bar(scores_long.iloc[1:,],
              x="variable",
              y="score",
              coloe="variable"
              title='Model score',
              )
# fig.show()
fig.write_html("../../pydata_talk/data/fig.html")
```

```python
fig = px.line(scores_long[scores_long["variable"]!= 'score'],
              x="id",
              y="score",
              color="variable",
              title='Classes individual scores',
              markers=True)
fig.show()
fig.write_html("../../pydata_talk/data/fig2.html")
```
